```python
import requests

def check_ollama_gpu():
    url = "http://localhost:11434/api/ps"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            models = data.get('models', [])
            
            if not models:
                print("Status: No models are currently loaded in Ollama.")
                print("Action: Run a model first (e.g., 'ollama run llama3') to check its resource usage.")
                return

            for model in models:
                name = model.get('name')
                # The 'processor' field explicitly shows GPU vs CPU usage
                processor = model.get('details', {}).get('processor', 'Unknown')
                
                print(f"Model: {name}")
                print(f"Detected Processor: {processor}")
                
                if "GPU" in processor.upper():
                    print("✅ Ollama is successfully using the NVIDIA GPU.")
                else:
                    print("⚠️ Ollama is running on the CPU.")
        else:
            print(f"Error: API returned status code {response.status_code}")
            
    except requests.exceptions.ConnectionError:
        print("Error: Could not connect to Ollama. Is the server running?")

if __name__ == "__main__":
    check_ollama_gpu()
```
